```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the inbuilt Titanic dataset from Seaborn
titanic = sns.load_dataset('titanic')

print("First 5 rows of the Titanic dataset:")
print(titanic.head())

# Display basic information about the dataset
print("\nDataset Information:")
print(titanic.info())
```

```python
sns.pairplot(titanic[['fare', 'age', 'pclass', 'survived']].dropna())
plt.suptitle("Pairplot of Selected Variables in the Titanic Dataset", y=1.02)
plt.show()

# Additionally, we can look at the distribution of the fare for each class.
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title("Boxplot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()

# ===============================
# Histogram of the 'fare' Variable
# ===============================
# Plot a histogram to see how the ticket price for each passenger is distributed.
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='fare', kde=True, bins=30)
plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```
