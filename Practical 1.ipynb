```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Student ID': [1, 2, 3, 4, 5],
    'Age': [18, 20, np.nan, 19, 22],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male'],
    'Exam Score': [90, 85, 75, 95, 65],
    'Study Hours': [4, 6, 8, np.nan, 5],
    'Attendance Percentage': [80, 95, 85, 70, 60]
}
df = pd.DataFrame(data)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df
```

```python
df['Study Hours']=df['Study Hours'].fillna(df['Study Hours'].median())
```

```python
df.info()
```

```python
from scipy.stats import zscore
df['z_score']=zscore(df['Exam Score'])
```

```python
outliers = df[df['z_score'].abs() > 3]
print(outliers)
```

```python

# Function to find outliers using IQR method
def find_outliers_IQR(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Checking for outliers in numerical columns
numerical_columns = ['Age', 'Exam Score', 'Study Hours', 'Attendance Percentage']

outliers = {}
for column in numerical_columns:
    outliers[column] = find_outliers_IQR(df, column)

# Print outliers for each numerical column
for column, outlier_data in outliers.items():
    print(f"Outliers in '{column}':")
    print(outlier_data[['Student ID', column]])
    print()

```

```python
for column, outlier_data in outliers.items():
    print(f"Outliers in '{column}':")
    print(outlier_data[['Student ID', column]])
    print("="*40)  # Separate output for clarity
```

```python
# Remove outliers for specific columns
df_cleaned = df[~df['Exam Score'].isin(outliers['Exam Score']['Exam Score'])]
df_cleaned = df_cleaned[~df_cleaned['Attendance Percentage'].isin(outliers['Attendance Percentage']['Attendance Percentage'])]
```
